```python
class Clock(object):
    """æ•°å­—æ—¶é’Ÿ"""

    def __init__(self, hour=0, minute=0, second=0):
        """åˆå§‹åŒ–æ–¹æ³•

        :param hour: æ—¶
        :param minute: åˆ†
        :param second: ç§’
        """
        self._hour = hour
        self._minute = minute
        self._second = second

    def run(self):
        """èµ°å­—"""
        self._second += 1
        if self._second == 60:
            self._second = 0
            self._minute += 1
            if self._minute == 60:
                self._minute = 0
                self._hour += 1
                if self._hour == 24:
                    self._hour = 0

    def __str__(self):
        """æ˜¾ç¤ºæ—¶é—´"""
        return '%02d:%02d:%02d' % \
               (self._hour, self._minute, self._second)


def main():
    clock = Clock(23, 59, 58)
    while True:
        print(clock)
        time.sleep(1)
        clock.run()


if __name__ == '__main__':
    main()
```
